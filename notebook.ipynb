```python
import pandas as pd
import numpy as np
from collections import Counter

DOSSIER = r"C:\Users\mhamd\Downloads"

# ═══════════════════════════════════════════
# CHARGEMENT DES DONNÉES
# ═══════════════════════════════════════════

df_core = pd.read_csv(DOSSIER + r"\watersec_core.csv", encoding="utf-8", low_memory=False)
df_gym  = pd.read_csv(DOSSIER + r"\watersec_gym.csv",  encoding="utf-8", low_memory=False)

# ═══════════════════════════════════════════
# PATTERN 1 : CONSOMMATION PAR MOMENT DE LA JOURNÉE
# ═══════════════════════════════════════════

def pattern_moment_journee(customer_id=None):
    df = df_core.copy()
    if customer_id:
        df = df[df["customer_id"] == customer_id]
    result = df.groupby("moment_journee")["consommation_ml"].agg(
        total="sum",
        moyenne="mean",
        nb_mesures="count"
    ).reset_index()
    result["total_litres"]   = (result["total"]   / 1000).round(2)
    result["moyenne_litres"] = (result["moyenne"] / 1000).round(2)
    result = result.sort_values("total_litres", ascending=False)
    return result[["moment_journee", "total_litres", "moyenne_litres", "nb_mesures"]]

# ═══════════════════════════════════════════
# PATTERN 2 : CONSOMMATION PAR JOUR DE LA SEMAINE
# ═══════════════════════════════════════════

def pattern_jour_semaine(customer_id=None):
    df = df_core.copy()
    if customer_id:
        df = df[df["customer_id"] == customer_id]
    result = df.groupby("jour")["consommation_ml"].agg(
        total="sum",
        moyenne="mean"
    ).reset_index()
    result["total_litres"]   = (result["total"]   / 1000).round(2)
    result["moyenne_litres"] = (result["moyenne"] / 1000).round(2)
    ordre = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
    result["ordre"] = result["jour"].map({j: i for i, j in enumerate(ordre)})
    result = result.sort_values("ordre")
    return result[["jour", "total_litres", "moyenne_litres"]]

# ═══════════════════════════════════════════
# PATTERN 3 : DÉTECTION ANOMALIES (Z-SCORE)
# ═══════════════════════════════════════════

def detecter_anomalies(customer_id=None, seuil=2.5):
    df = df_core.copy()
    if customer_id:
        df = df[df["customer_id"] == customer_id]
    df["mean"]   = df.groupby("device")["consommation_ml"].transform("mean")
    df["std"]    = df.groupby("device")["consommation_ml"].transform("std")
    df["zscore"] = (df["consommation_ml"] - df["mean"]) / df["std"].replace(0, 1)
    anomalies    = df[df["zscore"].abs() > seuil].copy()
    anomalies["consommation_litres"] = (anomalies["consommation_ml"] / 1000).round(2)
    return anomalies[["customer_id", "device", "consommation_litres",
                       "statut_alerte", "jour", "moment_journee", "zscore"]].head(20)

# ═══════════════════════════════════════════
# PATTERN 4 : SÉQUENCES FLUSH → SINK (RÉSIDENTIEL)
# ═══════════════════════════════════════════

def pattern_sequences_residentielles():
    df = df_core[df_core["customer_id"] == "customerC"].copy()
    if "sub_category_name" not in df.columns:
        return "Colonne sub_category_name absente pour customerC"
    sequences = []
    devices = df["device"].unique()
    for device in devices:
        df_device  = df[df["device"] == device].copy()
        categories = df_device["sub_category_name"].dropna().tolist()
        for i in range(len(categories) - 1):
            seq = f"{categories[i]} → {categories[i+1]}"
            sequences.append(seq)
    compteur      = Counter(sequences)
    top_sequences = pd.DataFrame(
        compteur.most_common(10),
        columns=["sequence", "occurrences"]
    )
    return top_sequences

# ═══════════════════════════════════════════
# PATTERN 5 : TENDANCE PAR SAISON
# ═══════════════════════════════════════════

def pattern_saison(customer_id=None):
    df = df_core.copy()
    if customer_id:
        df = df[df["customer_id"] == customer_id]
    result = df.groupby("saison")["consommation_ml"].agg(
        total="sum",
        moyenne="mean"
    ).reset_index()
    result["total_litres"]   = (result["total"]   / 1000).round(2)
    result["moyenne_litres"] = (result["moyenne"] / 1000).round(2)
    result = result.sort_values("total_litres", ascending=False)
    return result[["saison", "total_litres", "moyenne_litres"]]

# ═══════════════════════════════════════════
# PATTERN 6 : FRÉQUENTATION GYM PAR MOMENT
# ═══════════════════════════════════════════

def pattern_gym_frequentation():
    df = df_gym.copy()
    df["frequentation_gym"] = pd.to_numeric(df["frequentation_gym"], errors="coerce")
    result = df.groupby("moment_journee").agg(
        conso_totale_litres=("consommation_ml", lambda x: round(x.sum()/1000, 2)),
        freq_moyenne=("frequentation_gym", "mean"),
        nb_sessions=("consommation_ml", "count")
    ).reset_index()
    result["freq_moyenne"] = result["freq_moyenne"].round(1)
    return result.sort_values("conso_totale_litres", ascending=False)

# ═══════════════════════════════════════════
# PATTERN 7 : FUITES DÉTECTÉES
# ═══════════════════════════════════════════

def detecter_fuites():
    df_all = pd.concat([df_core, df_gym], ignore_index=True)
    fuites = df_all[df_all["statut_alerte"].str.contains("Fuite", na=False, case=False)]
    result = fuites.groupby(["customer_id", "device", "statut_alerte"]).agg(
        nb_occurrences=("consommation_ml", "count"),
        conso_totale_litres=("consommation_ml", lambda x: round(x.sum()/1000, 2))
    ).reset_index()
    return result.sort_values("nb_occurrences", ascending=False)

# ═══════════════════════════════════════════
# TEST DE TOUS LES PATTERNS
# ═══════════════════════════════════════════

print("=== PATTERN 1 : Moment de la journée ===")
print(pattern_moment_journee())

print("\n=== PATTERN 2 : Jour de la semaine ===")
print(pattern_jour_semaine())

print("\n=== PATTERN 3 : Anomalies ===")
print(detecter_anomalies())

print("\n=== PATTERN 4 : Séquences résidentielles ===")
print(pattern_sequences_residentielles())

print("\n=== PATTERN 5 : Saisons ===")
print(pattern_saison())

print("\n=== PATTERN 6 : Fréquentation Gym ===")
print(pattern_gym_frequentation())

print("\n=== PATTERN 7 : Fuites ===")
print(detecter_fuites())

print("\n✅ Moteur patterns prêt !")
```

```python
import requests
import pandas as pd
from datetime import datetime

OPENWEATHER_KEY = "5662feedf185c83f056260dcdb5f9a76"

# ═══════════════════════════════════════════
# MÉTÉO ACTUELLE — TUNIS
# ═══════════════════════════════════════════

def get_meteo_actuelle(ville="Tunis"):
    url = f"http://api.openweathermap.org/data/2.5/weather"
    params = {
        "q":     ville + ",TN",
        "appid": OPENWEATHER_KEY,
        "units": "metric",
        "lang":  "fr"
    }
    r = requests.get(url, params=params)
    data = r.json()
    if r.status_code != 200:
        return f"Erreur météo : {data.get('message', 'inconnue')}"
    return {
        "ville":       data["name"],
        "temperature": data["main"]["temp"],
        "ressenti":    data["main"]["feels_like"],
        "humidite":    data["main"]["humidity"],
        "description": data["weather"][0]["description"],
        "vent_kmh":    round(data["wind"]["speed"] * 3.6, 1)
    }

# ═══════════════════════════════════════════
# MÉTÉO 5 JOURS — TUNIS
# ═══════════════════════════════════════════

def get_meteo_5jours(ville="Tunis"):
    url = f"http://api.openweathermap.org/data/2.5/forecast"
    params = {
        "q":     ville + ",TN",
        "appid": OPENWEATHER_KEY,
        "units": "metric",
        "lang":  "fr"
    }
    r    = requests.get(url, params=params)
    data = r.json()
    if r.status_code != 200:
        return f"Erreur météo : {data.get('message', 'inconnue')}"

    previsions = []
    for item in data["list"]:
        previsions.append({
            "datetime":    item["dt_txt"],
            "temperature": item["main"]["temp"],
            "humidite":    item["main"]["humidity"],
            "description": item["weather"][0]["description"],
            "vent_kmh":    round(item["wind"]["speed"] * 3.6, 1)
        })
    return pd.DataFrame(previsions)

# ═══════════════════════════════════════════
# CALENDRIER TUNISIE 2024-2025
# ═══════════════════════════════════════════

CALENDRIER_TUNISIE = {
    # Ramadan
    "2024-03-11": "Début Ramadan 2024",
    "2024-04-09": "Fin Ramadan 2024",
    "2025-03-01": "Début Ramadan 2025",
    "2025-03-30": "Fin Ramadan 2025",

    # Aïd
    "2024-04-10": "Aïd El Fitr 2024",
    "2024-06-16": "Aïd El Adha 2024",
    "2025-03-31": "Aïd El Fitr 2025",
    "2025-06-06": "Aïd El Adha 2025",

    # Fêtes nationales
    "2024-01-14": "Fête de la Révolution",
    "2024-03-20": "Fête de l'Indépendance",
    "2024-04-09": "Fête des Martyrs",
    "2024-05-01": "Fête du Travail",
    "2024-07-25": "Fête de la République",
    "2024-08-13": "Fête de la Femme",
    "2024-10-15": "Fête de l'Evacuation",

    # Vacances scolaires
    "2024-01-01": "Vacances scolaires - Nouvel An",
    "2024-12-25": "Vacances scolaires - Hiver",
}

def get_evenement_tunisie(date_str):
    return CALENDRIER_TUNISIE.get(date_str, "Jour normal")

def enrichir_avec_contexte(date_str):
    meteo   = get_meteo_actuelle()
    evenement = get_evenement_tunisie(date_str)
    return {
        "date":      date_str,
        "evenement": evenement,
        "meteo":     meteo
    }

# ═══════════════════════════════════════════
# TEST
# ═══════════════════════════════════════════

print("=== MÉTÉO ACTUELLE ===")
print(get_meteo_actuelle())

print("\n=== MÉTÉO 5 JOURS ===")
print(get_meteo_5jours().head())

print("\n=== CALENDRIER TUNISIE ===")
today = datetime.now().strftime("%Y-%m-%d")
print(enrichir_avec_contexte(today))

print("\n✅ Météo + Calendrier Tunisie prêts !")
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

DOSSIER = r"C:\Users\mhamd\Downloads"

df_core = pd.read_csv(DOSSIER + r"\watersec_core.csv", encoding="utf-8", low_memory=False)
df_gym  = pd.read_csv(DOSSIER + r"\watersec_gym.csv",  encoding="utf-8", low_memory=False)

# ═══════════════════════════════════════════
# GRAPHIQUE 1 : CONSOMMATION PAR CLIENT
# ═══════════════════════════════════════════

def graph_conso_par_client():
    df_all = pd.concat([df_core, df_gym], ignore_index=True)
    result = df_all.groupby("customer_id")["consommation_ml"].sum().reset_index()
    result["consommation_litres"] = (result["consommation_ml"] / 1000).round(2)

    fig = px.bar(
        result,
        x="customer_id",
        y="consommation_litres",
        title="Consommation totale par client (litres)",
        color="customer_id",
        text="consommation_litres",
        color_discrete_sequence=px.colors.qualitative.Bold
    )
    fig.update_traces(texttemplate="%{text:.0f} L", textposition="outside")
    fig.update_layout(showlegend=False, xaxis_title="Client", yaxis_title="Litres")
    fig.show()

# ═══════════════════════════════════════════
# GRAPHIQUE 2 : TENDANCE PAR MOMENT DE LA JOURNÉE
# ═══════════════════════════════════════════

def graph_moment_journee(customer_id=None):
    df = df_core.copy()
    if customer_id:
        df = df[df["customer_id"] == customer_id]
    result = df.groupby("moment_journee")["consommation_ml"].mean().reset_index()
    result["moyenne_litres"] = (result["consommation_ml"] / 1000).round(2)

    ordre = ["Matin", "Après-midi", "Soirée", "Nuit"]
    result["ordre"] = result["moment_journee"].map({j: i for i, j in enumerate(ordre)})
    result = result.sort_values("ordre")

    fig = px.bar(
        result,
        x="moment_journee",
        y="moyenne_litres",
        title=f"Consommation moyenne par moment de la journée {'- ' + customer_id if customer_id else ''}",
        color="moment_journee",
        text="moyenne_litres",
        color_discrete_sequence=px.colors.qualitative.Pastel
    )
    fig.update_traces(texttemplate="%{text:.2f} L", textposition="outside")
    fig.update_layout(showlegend=False, xaxis_title="Moment", yaxis_title="Litres")
    fig.show()

# ═══════════════════════════════════════════
# GRAPHIQUE 3 : CONSOMMATION PAR SAISON
# ═══════════════════════════════════════════

def graph_saison(customer_id=None):
    df = df_core.copy()
    if customer_id:
        df = df[df["customer_id"] == customer_id]
    result = df.groupby("saison")["consommation_ml"].sum().reset_index()
    result["total_litres"] = (result["consommation_ml"] / 1000).round(2)

    fig = px.pie(
        result,
        names="saison",
        values="total_litres",
        title=f"Répartition par saison {'- ' + customer_id if customer_id else ''}",
        color_discrete_sequence=px.colors.qualitative.Safe
    )
    fig.show()

# ═══════════════════════════════════════════
# GRAPHIQUE 4 : ANOMALIES PAR DEVICE
# ═══════════════════════════════════════════

def graph_anomalies(customer_id=None):
    df = df_core.copy()
    if customer_id:
        df = df[df["customer_id"] == customer_id]

    df["mean"]   = df.groupby("device")["consommation_ml"].transform("mean")
    df["std"]    = df.groupby("device")["consommation_ml"].transform("std")
    df["zscore"] = (df["consommation_ml"] - df["mean"]) / df["std"].replace(0, 1)
    df["statut"] = df["zscore"].abs().apply(lambda z: "Anomalie" if z > 2.5 else "Normal")
    df["conso_litres"] = (df["consommation_ml"] / 1000).round(2)

    fig = px.scatter(
        df.sample(min(5000, len(df))),
        x="device",
        y="conso_litres",
        color="statut",
        title=f"Anomalies de consommation {'- ' + customer_id if customer_id else ''}",
        color_discrete_map={"Normal": "#2196F3", "Anomalie": "#F44336"},
        opacity=0.6
    )
    fig.update_layout(xaxis_title="Device", yaxis_title="Litres")
    fig.show()

# ═══════════════════════════════════════════
# GRAPHIQUE 5 : COMPARAISON CABINES GYM
# ═══════════════════════════════════════════

def graph_comparaison_cabines():
    result = df_gym.groupby("device")["consommation_ml"].agg(
        total="sum",
        moyenne="mean"
    ).reset_index()
    result["total_litres"]   = (result["total"]   / 1000).round(2)
    result["moyenne_litres"] = (result["moyenne"] / 1000).round(2)

    fig = make_subplots(rows=1, cols=2,
                        subplot_titles=("Total par cabine", "Moyenne par cabine"))

    fig.add_trace(go.Bar(
        x=result["device"], y=result["total_litres"],
        name="Total", marker_color="#1E88E5",
        text=result["total_litres"],
        texttemplate="%{text:.0f} L"
    ), row=1, col=1)

    fig.add_trace(go.Bar(
        x=result["device"], y=result["moyenne_litres"],
        name="Moyenne", marker_color="#43A047",
        text=result["moyenne_litres"],
        texttemplate="%{text:.2f} L"
    ), row=1, col=2)

    fig.update_layout(title="Comparaison des cabines Gym", showlegend=False)
    fig.show()

# ═══════════════════════════════════════════
# GRAPHIQUE 6 : FRÉQUENTATION GYM VS CONSOMMATION
# ═══════════════════════════════════════════

def graph_gym_freq_vs_conso():
    df = df_gym.copy()
    df["frequentation_gym"] = pd.to_numeric(df["frequentation_gym"], errors="coerce")
    df["conso_litres"]      = (df["consommation_ml"] / 1000).round(2)
    df_clean = df.dropna(subset=["frequentation_gym"])

    if len(df_clean) == 0:
        print("⚠️ frequentation_gym vide — graphique par moment de la journée")
        result = df.groupby("moment_journee")["conso_litres"].mean().reset_index()
        fig = px.bar(
            result,
            x="moment_journee",
            y="conso_litres",
            title="Consommation moyenne Gym par moment de la journée",
            color="moment_journee",
            color_discrete_sequence=px.colors.qualitative.Bold
        )
        fig.show()
        return

    df_sample = df_clean.sample(min(2000, len(df_clean)))
    fig = px.scatter(
        df_sample,
        x="frequentation_gym",
        y="conso_litres",
        color="moment_journee",
        title="Fréquentation Gym vs Consommation",
        trendline="ols",
        labels={"frequentation_gym": "Fréquentation", "conso_litres": "Litres"}
    )
    fig.show()

# ═══════════════════════════════════════════
# TEST TOUS LES GRAPHIQUES
# ═══════════════════════════════════════════

graph_conso_par_client()
graph_moment_journee()
graph_saison()
graph_anomalies()
graph_comparaison_cabines()
graph_gym_freq_vs_conso()

print("✅ Graphiques Plotly prêts !")
```
